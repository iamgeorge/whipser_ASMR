```python
from audiocraft.models import musicgen
from audiocraft.utils.notebook import display_audio
import torch
```

```python
model = musicgen.MusicGen.get_pretrained('small', device='cuda')
model.set_generation_params(duration=8)
```

```python
res = model.generate([
    'tranquil, immersive ambient sounds blending soft ocean waves, distant seabird calls, and gentle harp melodies',
    'peaceful lo-fi beats with a steady, calming rhythm accompanied by subtle crackling fireplace sounds and soft piano accents',
    'meditative ambient music featuring delicate rain showers, low droning pads, and gentle bell tones',
    'smooth jazz-infused instrumental with mellow saxophone lines, slow drums, warm guitar strums, and a relaxed, soulful vibe',
    'minimalistic acoustic guitar melodies intertwined with gentle wind chimes, rustling leaves, and warm, earthy tones'
],
    progress=True
)
display_audio(res, 32000)
```
